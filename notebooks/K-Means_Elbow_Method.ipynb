# K-Means Elbow Method (Metrics)

## 1.0 Imports

```python
from sklearn import datasets as ds
from matplotlib import pyplot as plt
from sklearn import cluster as ct
from sklearn import metrics as mt
import numpy as np
```

## 2.0 Load datasets

```python
X, y = ds.make_moons(
    n_samples=1000,
    noise=0.1
)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

## 3.0 Training K-Means

```python
clusters = np.arange(2, 21, 1)

silhouette_list = []

for k in clusters:

    # define
    model = ct.KMeans(
        n_clusters=k,
        init="random",
        n_init=10,
        random_state=0
    )

    # fit (training)
    labels = model.fit_predict(X)

    # performance (evaluation)
    silhouette_avg = mt.silhouette_score(X, labels)
    silhouette_list.append(silhouette_avg)

    print(f"Cluster k: {k} - Silhouette Score Avg: {silhouette_avg}")
```

```python
plt.plot(clusters, silhouette_list, marker="o")
plt.xlabel("X label")
plt.ylabel("Silhouette Score Avg")
```

## 4.0 Fine tunning

```python
k = 10

model = ct.KMeans(
    n_clusters=k,
    init="random",
    n_init=k,
    random_state=0
)

# fit
labels = model.fit_predict(X)

# performance
silhouette_avg = mt.silhouette_score(X, labels)
silhouette_list.append(silhouette_avg)

plt.scatter(X[:, 0], X[:, 1], c=labels)
```
