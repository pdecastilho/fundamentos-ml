# Decision Tree

## 1.0 Imports

```python
import cv2
from sklearn import datasets as dt
from sklearn import tree as tr
from matplotlib import pyplot as plt
```

## 2.0 Load dataset

```python
iris = dt.load_iris()

X = iris.data[:, 2:]
y = iris.target
```

```python
# define

# max_depth: controle do tamanho máximo de quebras da árvore, ou seja, o tamanho máximo de crescimento da árvore, ou a quantidade de recortes do espaço

# min_samples_leaf: o número mínimo de amostras que a folha deve ter antes de fazer uma separação e gerar nós filhos

# min_samples_split: o número mínimo de amostras que a folha deve ter após fazer uma nova separação

# max_features: o número máximo de atributos para avaliar a divisão de cada nó
model_tree = tr.DecisionTreeClassifier(
    max_depth=10,
    min_samples_leaf=20
)

# fit (training)
model_tree.fit(X, y)

# predict
```

```python
tr.export_graphviz(
    model_tree,
    out_file="tree.dot",
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
```
